```python
#importando as bibliotecas
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
```

```python
def monta_dataframe(pasta):
  '''
  Retorna um Pandas DataFrame com as informações do dataset e uma lista com imagens repetidas
  Parâmetro:
  pasta (str) : Informe o caminho até a pasta 'Dataset_BUSI_with_GT'
  Exemplo: /content/dataset_projeto/Dataset_BUSI_with_GT/
  '''
  fotos = sorted(glob(f"{pasta}*/*).png"))
  lista_total = []
  lista_repetidos = []
  for caminho in fotos:
    mascaras = sorted(glob(caminho[:-4] + '*_mask*'))
    caminho_separado = caminho.split('/')
    cont=0
    for caminho_masc in mascaras:
      cont+=cont+1
      lista_indiv = [caminho_separado[5]]
      lista_indiv.append(caminho_separado[4])
      lista_indiv.append(caminho)
      lista_indiv.append(caminho_masc)
      lista_total.append(lista_indiv)
      if cont > 1:
        lista_repetidos.append(caminho_separado[5])
  return pd.DataFrame(lista_total, columns=['Nome','Categoria','Original','Ground Truth']), lista_repetidos
```

```python
dataframe, repetidos = monta_dataframe('/content/dataset_projeto/Dataset_BUSI_with_GT/')
dataframe
```
